```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax= plt.subplots()
```

```python
fig,(ax1,ax2)= plt.subplots(nrows=2)
```

```python
fig,(ax1,ax2)= plt.subplots(ncols=2)
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2)
```

```python
fig,ax= plt.subplots(figsize=(3,3))
```

```python
data= [1,2,3,4,5,6,8,10,12]
fig,ax= plt.subplots(figsize=(3,3))
ax.set_xlabel('No of Element')
ax.set_ylabel('Y- value')
ax.plot(data)
```

```python
data= [1,2,3,4,5,6,8,10,12]
fig,ax= plt.subplots(figsize=(3,3))
ax.set_xlabel('No of Element', fontdict={'fontsize':10,'color':'green'})
ax.set_ylabel('Y- value')
ax.plot(data)
```

```python
data= [1,2,3,4,5,6,8,10,12]
fig,ax= plt.subplots(figsize=(10,4))
ax.set_xlabel('No of Element', fontdict={'fontsize':10,'color':'green'})
ax.set_ylabel('Y- value')
ax.grid(True)
ax.set_title('Basic of visualisation')
ax.plot(data,marker='o',markersize=6,markerfacecolor='yellow')
```

```python
data=[1,2,3,4,5,6,7,8,10,12]
data2=[22,23,34,55,7,23,24]
fig,ax= plt.subplots(figsize=(10,4))
ax.plot(data,marker='x', markersize=11,markerfacecolor='red',linestyle=':',alpha=0.5)
ax.plot(data2,marker='x', markersize=11,markerfacecolor='red',linestyle='--',color='green')
```

```python
data=[1,2,3,4,5,6,7,8,10,12]
data2=[22,23,34,55,7,23,24]
fig,(ax1,ax2)= plt.subplots(figsize=(10,4),ncols=2)
ax1.plot(data,marker='x', markersize=11,markerfacecolor='red',linestyle=':',alpha=0.5)
ax2.plot(data2,marker='x', markersize=11,markerfacecolor='red',linestyle='--',color='green')
```
